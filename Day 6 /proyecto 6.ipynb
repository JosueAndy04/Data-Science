```python
valo_cero = {'Oro': 0, 'Plata':0, 'Bronce':0 }
df = df.fillna(valo_cero)
df['Oro'] = df['Oro'].astype(int)
df['Plata'] = df['Plata'].astype(int)
df['Bronce'] = df['Bronce'].astype(int)

df.head()
```
