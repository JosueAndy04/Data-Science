## Dashboard Analisis Exploratorio de Datos

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

1.- Cargar el dataset diamonds de Seaborn

```python
diamantes = sns.load_dataset('diamonds')
```

3.- Visualizaciones

```python
sns.set_theme(style='dark', palette='deep')
```

3.1 Relacion numerica (peso vs precio quitlates)

```python

p_q = sns.relplot(data=diamantes, x='carat', y='price', style='cut', hue='cut')
p_q.set_axis_labels('Peso quilate', 'Precio');
```

3.2 Distribucion de variables numericas (precio, peso, en quilates)

```python
p_q_d = sns.displot(data=diamantes, x='price', kind='hist', kde=True, height=4, aspect=2)
p_q_d.figure.suptitle('Distribucion del precio del diamante', va='baseline', ha='center')
p_q_d.set_axis_labels('Precio', 'Frecuencia');
```

```python
his = sns.histplot(data=diamantes, x='carat', kde=True)
his.set_title('Distribucion del precio del diamante')
his.set(xlabel='Peso', ylabel='Frecuencia');
```

3.3 Comparaciones de variables numericas por categoria (calida del corte)

```python
p_q_d = sns.catplot(data=diamantes, kind='bar', x='cut',y='price', hue='color', height=5, aspect=2)
p_q_d.figure.suptitle('Comparacion del precio de los diamantes por la calidad de corte', va='baseline', ha='center')
p_q_d.set_axis_labels('Corte', 'Precio');
```

4.- Utilizar joinplot para explorar relaciones especificas entre pares de variables, incluyendo informacion estadistica adicional

```python
unido = sns.jointplot(data=diamantes, x='table', y='price', kind='kde', fill=True, space=0)
unido.figure.suptitle('Relacion entre el precio y la tabla del diamante', va='baseline', ha='center')
unido.set_axis_labels('Tabla', 'Precio');
```

5.- Emplear pairplot para visualizar las relaciones entre todas las variables numeicas del dataset de manera conjunta

```python
par = sns.pairplot(data=diamantes, hue='cut', corner=True)
par.map_lower(sns.scatterplot)
```
