## Analizador de ventas

```python
import pandas as pd
```

1.- Lectura de datos

```python
ruta = '/home/anderson-josue/Documents/Data Science/Day 7 /Datos_Ventas_Tienda.csv'
ruta1 = '/home/anderson-josue/Documents/Data Science/Day 7 /Datos_Ventas_Tienda2.csv'

df = pd.read_csv(ruta)
df1 = pd.read_csv(ruta1)
```

```python
df
```

```python
df1
```

2.- Fusion de datos: concatenar los dos DataFrame

```python
ventas = pd.concat([df, df1], ignore_index=True)
ventas
```

3.- Limpieza de datos, filtrado y transformaciones basicas


```python
ventas.describe()
```

```python
ventas.info()
```

```python
ventas.sort_values(by='Total Venta', ascending=False)
```

4.- Analisis de ventas: Producto mas vendido, Mes con mas ventas

Producto mas vendido

```python
producto = ventas.groupby('Producto')['Cantidad'].sum().sort_values(ascending=False).head(1)
producto
```

Mes con mas ventas

```python
ventas['Fecha'] = pd.to_datetime(ventas['Fecha'], format='%m/%d/%Y')
ventas
```

```python
ventas['Meses'] = ventas['Fecha'].dt.month
ventas
```

```python
mes = pd.Series(['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
                 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'])
ventas['Meses'] = ventas['Meses'].apply(lambda x: mes[x - 1])
ventas
```

```python
mes = ventas.groupby('Meses')['Total Venta'].sum().sort_values(ascending=False).head(1)
mes
```

5.- Datos agrupados: Categoria y analiza las ventas

```python
categoria = ventas.groupby('Producto')['Total Venta'].sum().sort_values(ascending=False)
categoria
```
